```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#数据读入
data1 = pd.read_csv("../Qt/Test_for_parkensonpen/savedata/2023_9_8_13_11_9.txt", names=['axis', 'time', 'attitude'], sep=" ")
```

```python
pitch = data1[data1['axis'] == 0]
pitch = pitch.drop(['axis'], axis=1)
print(pitch)
```

```python
x = pitch['time']
y = pitch['attitude']
plt.plot(x, y, label="pitch")
plt.xlabel("time")
plt.ylabel("attitude")
plt.title("Time domain map")
plt.legend()
plt.show()
```
